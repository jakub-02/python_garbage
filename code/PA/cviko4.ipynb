```python
import pandas as pd
```

```python
ba = pd.read_csv('bratislava.csv')
ke = pd.read_csv('data/kosice.csv')
ni = pd.read_csv('data/nitra.csv')
os = pd.read_csv('data/ostrava.csv')
br = pd.read_csv('data/brno.csv')
```

```python
ba.head()
```

```python
os.head()
```

```python
print(ba.produkt.unique())
print(ba.cena.unique())

print(os.produkt.unique())
print(os.cena.unique())
```

```python
os['produkt'] = os['produkt'].replace('Chleb', 'Chlieb')
os['produkt'] = os['produkt'].replace('Svetr', 'Sveter')
os['produkt'] = os['produkt'].replace('Triko', 'Tricko')
os['produkt'] = os['produkt'].replace('Mleko', 'Mlieko')
os.head()
```

```python
br['produkt'] = br['produkt'].replace('Chleb', 'Chlieb')
br['produkt'] = br['produkt'].replace('Svetr', 'Sveter')
br['produkt'] = br['produkt'].replace('Triko', 'Tricko')
br['produkt'] = br['produkt'].replace('Mleko', 'Mlieko')
br.head()
```

```python
# aktuanlny kurz, zdroj google
kurz = 25.26
# [   40   250    55   620 14500  6300  9200]
br['cena'] = br['cena'].replace(40, 40/kurz)
br['cena'] = br['cena'].replace(250, 250/kurz)
br['cena'] = br['cena'].replace(55, 55/kurz)
br['cena'] = br['cena'].replace(620, 620/kurz)
br['cena'] = br['cena'].replace(14500, 14500/kurz)
br['cena'] = br['cena'].replace(6300, 6300/kurz)
br['cena'] = br['cena'].replace(9200, 9200/kurz)

br.head()
```

```python
os['cena'] = os['cena'].replace(40, 40/kurz)
os['cena'] = os['cena'].replace(250, 250/kurz)
os['cena'] = os['cena'].replace(55, 55/kurz)
os['cena'] = os['cena'].replace(620, 620/kurz)
os['cena'] = os['cena'].replace(14500, 14500/kurz)
os['cena'] = os['cena'].replace(6300, 6300/kurz)
os['cena'] = os['cena'].replace(9200, 9200/kurz)
```

```python
os.head()
```

```python
ba['Mesto'] = 'bratislava'
ke['Mesto'] = 'kosice'
ni['Mesto'] = 'nitra'
os['Mesto'] = 'ostrava'
br['Mesto'] = 'brno'

ba['Krajina'] = 'sk'
ke['Krajina'] = 'sk'
ni['Krajina'] = 'sk'
os['Krajina'] = 'cz'
br['Krajina'] = 'cz'
```

```python
predaje = pd.concat([ba, ke, ni, os, br], ignore_index=True)
```

```python
predaje.head()
```

```python
predaje['cas_nakupu'] = pd.to_datetime(predaje['cas_nakupu'])

predaje['kvartal'] = predaje['cas_nakupu'].dt.to_period('Q').astype(str)
predaje.head()
```

```python
predaje['mesiac'] = predaje['cas_nakupu'].dt.month
predaje.head()
```

```python
df = predaje.drop(columns=['cas_nakupu'])
```

```python
df.head()
```

```python
# mlieko, chlieb -- potraviny
# tricko, sveter -- oblecenie
# mobil, tablet, laptop -- elektro

category_mapping = {
    'Mlieko': 'potraviny',
    'Chlieb': 'potraviny',
    'Tricko': 'oblecenie',
    'Sveter': 'oblecenie',
    'Mobil': 'elektro',
    'Tablet': 'elektro',
    'Laptop': 'elektro'
}

df['kategoria'] = df['produkt'].map(category_mapping)
df.head()
```
